## Sine Wave

Nice and quiet sine waves.


```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

asin = np.sin(3*2*np.pi*(np.arange(1000))/1000)
acos = np.cos(3*2*np.pi*(np.arange(1000))/1000)

plt.plot(asin)
plt.plot(acos)
```
